```python
#  Import and read the charity_data.csv.
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import LeakyReLU, Dense
from tensorflow.keras.activations import elu


import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(["EIN"],1)

#dropping status since they are all 1
application_df = application_df.drop(["STATUS"],1)
```

```python
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
print(application_cat)
```

```python
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[application_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_cat)
encode_df.head()
```

```python
application_df = application_df.merge(encode_df,left_index=True, right_index=True)

#application_df = application_df.drop(application_cat)
print(application_df.info())

# application2_df=pd.merge(application_df, encodetmp_CLASSIFICATION,left_index=True,right_index=True)
# application3_df=pd.merge(application2_df, encodetmp_APPLICATION_TYPE, left_index=True ,right_index=True)

# application4_df=application3_df.drop(columns='APPLICATION_TYPE')
# application5_df=application4_df.drop(columns='CLASSIFICATION')


# print(application5_df)
#application_df.merge(encodetmp_APPLICATION_TYPE,left_index=True,right_index=True)


application_df = application_df.drop(application_cat,1)
application_df.head()
```

```python
X = application_df.drop(["IS_SUCCESSFUL"], axis='columns').values

y = application_df["IS_SUCCESSFUL"].values

#Dropping 
print(application_df)

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
